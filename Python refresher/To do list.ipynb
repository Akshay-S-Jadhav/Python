# To do list

This is the program that enables user to manage their to do list

```python
def to_do_list():
    tasks = []
    
    while True:
        
        print("\n 1. Add task \n 2. View tasks \n 3. Mark task as done \n 4. Exit \n 5. Delete list")
        choice = input("Enter your choice: ")
        
        if choice == "1":
            task = input("Enter task you want to do : ")
            tasks.append(task)
            print("Task Added successfully")
            while True:
                more = input("Do you want to add more tasks, if yes type 'y' else 'n' : ")
                if more == 'y':
                    task = input("Enter task you want to do : ")
                    tasks.append(task)
                    print("Task Added successfully")
                else:
                    break
        elif choice == "2":
            if tasks:
                print("\n Your tasks:")
                for index, task in enumerate(tasks, start=1):
                    print(f"{index}. {task}")
            else:
                print("No tasks.")
        elif choice == "3":
            if tasks:
                task_index = int(input("Enter task number to mark as done: ")) - 1
                if 0 <= task_index < len(tasks):
                    print(f"Marked '{tasks[task_index]}' as done.")
                    del tasks[task_index]
                else:
                    print("Invalid task number.")
                    
            else:
                print("No tasks.")
        elif choice == "4":
            break
        elif choice == "5":
            del tasks[:]
            print("List is deleted ")
        else:
            print("Invalid Choice. Please try again. ")
            
to_do_list()
```
